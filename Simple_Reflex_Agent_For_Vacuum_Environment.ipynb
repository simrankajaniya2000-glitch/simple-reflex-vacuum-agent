```python
# Simple Reflex Agent for Vacuum Environment
# Author: Simran Kajaniya
# Date: January 27, 2025
# Purpose: Implement a simple reflex agent to clean a two-square environment.

def simple_reflex_agent(environment, initial_position):
    """Simulate the behavior of a simple reflex agent."""
    score = 0

    position = initial_position

    for _ in range(2):  # 2 moves for two squares A and B
      # Check if the current square is dirty
        if environment[position] == "dirty":
            print(f"Square {position} is dirty. Cleaning...") # Clean the square
            environment[position] = "clean"  # Mark the square as clean
            score += 1  # Add 1 point to the performance score

# Move to the next square
        position = "A" if position == "B" else "B"
        print(f"Moving to square {position}...")

    return score

def main():
    """Main function to test the simple reflex agent."""

# All possible initial configurations
    configurations = [
        {"A": "clean", "B": "clean"},
        {"A": "dirty", "B": "clean"},
        {"A": "clean", "B": "dirty"},
        {"A": "dirty", "B": "dirty"},
    ]

    for config in configurations:
        print("\nTesting configuration:", config)
        for start_position in ["A", "B"]:
            print(f"Starting at square {start_position}...")

            environment = config.copy()

            score = simple_reflex_agent(environment, start_position)
            print(f"Final environment: {environment}")
            print(f"Performance score: {score}")

if __name__ == "__main__":
    main()
```
